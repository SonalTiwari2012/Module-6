```python
import  numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

z=np.linspace(20,-10,10)
y=sigmoid(z)

plt.plot(z,y,label="sigmoid",color='Blue')
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.show()
```

```python
#Tanh- Hyperbolic Tangent
def sigmoid(z):
    return np.tanh(z)

z=np.linspace(30,-5,10)
y=sigmoid(z)

plt.plot(z,y,label="Sigmoid",color='red')
plt.title("Sigmoid Activation Function")
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

```python
#Relu - Rectified Liner Unit
def relu(z):
    return np.maximum(0,z)

z=np.linspace(-3,1,2)
y=relu(z)

plt.plot(z,y,label="Rectified Liner Unit (ReLu)",color='red')
plt.title("Relu Activation Function")
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```
