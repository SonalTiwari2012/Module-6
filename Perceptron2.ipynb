```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron 
from mlxtend.plotting import plot_decision_regions 
```

```python
df=pd.read_csv('Customer.csv') 
df
```

```python
df.isnull().sum() 
```

```python
df.columns
```

```python
#Encode categorical data
from sklearn.preprocessing import LabelEncoder 
```

```python
label_encoders={} 
for column in df.select_dtypes(include=['object']).columns:
    le=LabelEncoder() 
    df[column]=le.fit_transform(df[column]) 
    label_encoders[column]=le 
```

```python
#feature selection 
x=df.drop(columns=['Churn']) 
y=df['Churn'] 
```

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report 
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) 
```

```python
#SLP- Single Layer Perceptron
slp= Sequential([Dense(1,activation='sigmoid',input_shape=(X_train.shape[1],))]) 
```

```python
#compiling
slp.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) 
```

```python
slp.fit(X_train,y_train,epochs=15,batch_size=16,validation_data=(X_test,y_test)) 
```

```python
prediction=(slp.predict(X_test)>0.5) 
```

```python
print(classification_report(y_test,prediction)) 
```

```python
#Define the MLP model
mlp_model=Sequential([ 
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')])  
```

```python
#Compile the model 
mlp_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Train the model 
mlp_model.fit(X_train,y_train,epochs=20,batch_size=32,validation_data=(X_test,y_test)) 
```

```python
#evaluate the model 
y_pred_mlp=(mlp_model.predict(X_test)>0.5).astype("int32") 
print(classification_report(y_test,y_pred_mlp)) 
```
